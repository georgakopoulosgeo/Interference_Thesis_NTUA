```python
df_cpu = df[df['Interference_Category'] == 'CPU'].copy()

print(df_cpu.shape)
df_cpu.head(100)
```

```python
import plotly.express as px

# 1. Filter for your specific RPS values
target_rps = [100, 3500]
plot_df = df_cpu[df_cpu['Given_RPS'].isin(target_rps)].copy()

# 2. Sort to ensure the axes look logical
plot_df = plot_df.sort_values(['Scenario_Label', 'Given_RPS'])

# 3. Create the 3D Scatter plot
fig = px.scatter_3d(
    plot_df, 
    x='Scenario_Label', 
    y='Given_RPS', 
    z='norm_perf',
    color='norm_perf',  # Color by performance for better visual depth
    title='Performance Impact: CPU Scenarios vs RPS',
    labels={'norm_perf': 'Normalized Perf', 'Given_RPS': 'RPS'}
)

fig.show()
```

```python
# 1. Get unique values for both RPS and Replicas
unique_rps = df_cpu['Given_RPS'].unique()
unique_replicas = df_cpu['Replicas'].unique()

# 2. Create "Isolation" rows for EVERY combination of RPS and Replicas
isolation_rows = []
for rps in unique_rps:
    for rep in unique_replicas:
        isolation_rows.append({
            'Scenario_Label': 'Isolation',
            'Given_RPS': rps,
            'Replicas': rep,
            'norm_perf': 1.0,
            'Interference_Category': 'CPU'
        })

df_isolation = pd.DataFrame(isolation_rows)

# 3. Combine
df_cpu_isolation = pd.concat([df_cpu, df_isolation], ignore_index=True)
```

```python
import plotly.graph_objects as go
import pandas as pd

# 1. Filter and sort
target_rps = [500, 1000, 1500, 2000]
plot_df = df_cpu_isolation[df_cpu_isolation['Given_RPS'].isin(target_rps)].copy()

# 2. Pivot the data into a grid format
# This creates a matrix where Z values are mapped to X and Y coordinates
pivot_df = plot_df.pivot_table(
    index='Given_RPS', 
    columns='Scenario_Label', 
    values='norm_perf', 
    aggfunc='mean'
)
column_order = ['Isolation', 'CPU1', 'CPU2', 'CPU3', 'CPU4']
# Filter to only columns that actually exist in pivot_df
existing_columns = [c for c in column_order if c in pivot_df.columns]
pivot_df = pivot_df[existing_columns]

# 3. Create the Surface
#fig = go.Figure(data=[go.Surface(
##    z=pivot_df.values, 
##    x=pivot_df.columns, 
#    y=pivot_df.index,
#    colorscale='Viridis',
#    colorbar=dict(
#        title='Norm Perf',
#        tickfont=dict(size=16)
#    ),
#    cmin=0,
#    cmax=1
#)])
fig = go.Figure()

# Define a simple color mapping or list for different surfaces
colorscales = ['Viridis', 'Plasma', 'Blues']
replica_counts = sorted(plot_df['Replicas'].unique())

for i, rep in enumerate(replica_counts):
    rep_df = plot_df[plot_df['Replicas'] == rep]
    
    pivot_rep = rep_df.pivot_table(
        index='Given_RPS', 
        columns='Scenario_Label', 
        values='norm_perf', 
        aggfunc='mean'
    )[existing_columns]

    fig.add_trace(go.Surface(
        z=pivot_rep.values,
        x=pivot_rep.columns,
        y=pivot_rep.index,
        name=f'{rep} Replicas',
        colorscale=colorscales[i % len(colorscales)],
        opacity=0.7,
        showscale=False # Keeps the legend cleaner
    ))

fig.update_layout(
    title=dict(text='CPU Performance Landscape', font=dict(size=30), x=0.5),
    width=1500,
    height=1200,
    scene=dict(
        xaxis=dict(
            # Adding <br> BEFORE the text pushes it DOWN/AWAY
            title=dict(text='<br><br>Interference Scenarios', font=dict(size=20)),
            tickfont=dict(size=16)
        ),
        yaxis=dict(
            # Adding <br> AFTER the text pushes it SIDEWAYS/AWAY
            title=dict(text='Traffic Load (rps)<br><br>', font=dict(size=20)),
            tickfont=dict(size=16)
        ),
        zaxis=dict(
            # For the Z-axis, leading breaks push it LEFT
            title=dict(text='Normalized Performance<br><br>', font=dict(size=20)),
            range=[0.1, 1],
            tickfont=dict(size=16)
        ),
        aspectmode='manual',
        aspectratio=dict(x=1, y=1, z=0.7)
    ),
    # You MUST increase margins if you use line breaks, or the text will vanish
    margin=dict(l=150, r=150, b=150, t=100) 
)
fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# 1. Setup the Subplot Grid (2x2)
replica_counts = sorted(plot_df['Replicas'].unique())[:4] # Limit to first 4 for 2x2
rows = 2
cols = 2

fig = make_subplots(
    rows=rows, cols=cols,
    specs=[[{'type': 'surface'}, {'type': 'surface'}],
           [{'type': 'surface'}, {'type': 'surface'}]],
    #subplot_titles=[f"{r} Replicas" for r in replica_counts],
    vertical_spacing=0.001,
    horizontal_spacing=0.001
)

# 2. Add a surface to each subplot
for i, rep in enumerate(replica_counts):
    # Calculate row and column (Plotly indices start at 1)
    curr_row = (i // cols) + 1
    curr_col = (i % cols) + 1
    
    rep_df = plot_df[plot_df['Replicas'] == rep]
    
    # Pivot for the specific replica
    pivot_rep = rep_df.pivot_table(
        index='Given_RPS', 
        columns='Scenario_Label', 
        values='norm_perf', 
        aggfunc='mean'
    ).reindex(columns=column_order).dropna(axis=1, how='all')

    fig.add_trace(
        go.Surface(
            z=pivot_rep.values,
            x=pivot_rep.columns,
            y=pivot_rep.index,
            colorscale='Viridis',
            cmin=0, cmax=1,
            name=f"Rep {rep}",
            showscale=(i == len(replica_counts) - 1) # Show colorbar only once
        ),
        row=curr_row, col=curr_col
    )

# 3. Update Layout
fig.update_layout(
    title_text="Performance Landscapes: 2x2 Replica Comparison",
    height=1000, 
    width=1200,
    scene=dict(xaxis_title='Interference', yaxis_title='RPS', zaxis_title='NPS'),
    # Note: For multiple subplots, you may need to update each scene (scene, scene2, etc.)
)

# Optional: Apply consistent axis labels across all scenes
for i in range(1, len(replica_counts) + 1):
    scene_name = f"scene{i}" if i > 1 else "scene"
    fig.update_layout({
        scene_name: dict(
            xaxis_title='Interference',
            yaxis_title='RPS',
            zaxis=dict(title='NPS', range=[0, 1])
        )
    })

fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# 1. Map Scenarios to numeric values for the Z-axis
# Using the standard order: Isolation (0) to CPU4 (4)
scenario_map = {'Isolation': 0, 'CPU1': 1, 'CPU2': 2, 'CPU3': 3, 'CPU4': 4}
plot_df['Z_Interference'] = plot_df['Scenario_Label'].map(scenario_map)

# 2. Filter out any NaNs in our plotting columns to ensure the plot isn't "empty"
plot_df = plot_df.dropna(subset=['Given_RPS', 'Replicas', 'Z_Interference', 'norm_perf'])

# 3. Create the 4D Visualization
# We use go.Isosurface or go.Scatter3d depending on data density.
# For typical test data, Scatter3d with large markers is the most reliable way 
# to see the "Volume" without interpolation errors.
fig = go.Figure(data=[go.Scatter3d(
    x=plot_df['Given_RPS'],
    y=plot_df['Replicas'],
    z=plot_df['Z_Interference'],
    mode='markers',
    marker=dict(
        size=12,
        color=plot_df['norm_perf'],   # The 4th Dimension (NPS)
        colorscale='RdYlGn',          # Green = High Perf, Red = Bad
        colorbar=dict(title="NPS", thickness=20),
        opacity=0.9,
        showscale=True,
        cmin=0,
        cmax=1
    )
)])

# 4. Update the layout to make the axes readable
fig.update_layout(
    title=dict(text='Performance Volume: Load vs Replicas vs Interference', x=0.5),
    width=1200,
    height=900,
    scene=dict(
        xaxis_title='Traffic Load (RPS)',
        yaxis_title='App Replicas',
        zaxis=dict(
            title='Interference Scenario',
            tickvals=list(scenario_map.values()),
            ticktext=list(scenario_map.keys())
        ),
        # Adjusting the camera so you can see the depth clearly
        camera=dict(
            eye=dict(x=1.5, y=1.5, z=1.2)
        )
    ),
    margin=dict(l=0, r=0, b=0, t=50)
)

fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

# 1. Prepare Data
target_rps = [500, 1000, 1500, 2000]
plot_df = df_cpu_isolation[df_cpu_isolation['Given_RPS'].isin(target_rps)].copy()

# --- CUSTOMIZATION SECTION ---
# Define exactly which scenarios you want and in what order
# Example: Put CPU4 first, or move Isolation to the end
scenario_order = ['Isolation', 'CPU1', 'CPU2', 'CPU3', 'CPU4']

# Define exactly which replica counts you want and in what order
# Example: [4, 2, 1] to show scaling down instead of up
replica_order = sorted(plot_df['Replicas'].unique()) 

# This list comprehension creates the final X-axis sequence based on your choices above
combined_order = [f"{s} | R{r}" for s in replica_order  for r in scenario_order] 
# -----------------------------

# 2. Process Labels
plot_df['Scenario_Replica'] = plot_df['Scenario_Label'] + " | R" + plot_df['Replicas'].astype(str)

# 3. Pivot & Reindex
# Reindexing is key—it forces the DataFrame to match your 'combined_order' exactly
pivot_df = plot_df.pivot_table(
    index='Given_RPS', 
    columns='Scenario_Replica', 
    values='norm_perf',
    aggfunc='mean'
).reindex(columns=combined_order)

# 4. Plot
fig = go.Figure(data=[go.Surface(
    z=pivot_df.values,
    # Map the columns (labels) to a simple numeric range [0, 1, 2...] 
    # to ensure Plotly spaces them evenly on the X-axis
    x=list(range(len(pivot_df.columns))),
    y=pivot_df.index,
    colorscale='Viridis',
    colorbar=dict(title='NPS', tickfont=dict(size=14))
)])

fig.update_layout(
    title=dict(text='NPS Landscape: Custom Scenario Order', font=dict(size=24), x=0.5),
    scene=dict(
        xaxis=dict(
            title='Scenario | Replica Count',
            # Replace the numeric 0, 1, 2... with your custom labels
            tickvals=list(range(len(pivot_df.columns))),
            ticktext=pivot_df.columns,
            tickfont=dict(size=10)
        ),
        yaxis=dict(title='Traffic Load (RPS)'),
        zaxis=dict(title='NPS', range=[0, 1])
    ),
    width=1500,
    height=900,
    margin=dict(l=50, r=50, b=100, t=100)
)

fig.show()
```
